```python

import yaml
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the .txt file as a dictionary
file_path = "output.txt"  # Replace with your actual file path
with open(file_path, 'r') as file:
    data = yaml.safe_load(file)

# Step 2: Extract RGB data
rgb_data = data['rgb']['data']
height = data['rgb']['height']
width = data['rgb']['width']

# Convert flat RGB data into a 3D NumPy array
image_array = np.array(rgb_data, dtype=np.uint8).reshape((height, width, 3))

# Step 3: Display the image
plt.imshow(image_array)
plt.axis('off')  # Hide the axes for better visualization
plt.show()
```
