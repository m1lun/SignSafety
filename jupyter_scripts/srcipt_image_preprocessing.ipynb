```python
import cv2

def template_matching_and_crop(image_path, template_path, output_path):
    # Load the input image and template
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    template = cv2.imread(template_path, cv2.IMREAD_COLOR)

    if image is None or template is None:
        print("Error: Could not load the image or template.")
        return

    # Convert images to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray_template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

    # Get dimensions of the template
    template_height, template_width = gray_template.shape

    # Perform template matching
    result = cv2.matchTemplate(gray_image, gray_template, cv2.TM_CCOEFF_NORMED)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    # Get the top-left corner of the matching region
    top_left = max_loc

    # Calculate the bottom-right corner of the rectangle
    bottom_right = (top_left[0] + template_width, top_left[1] + template_height)

    # Draw a rectangle around the matched region
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)

    # Crop the matching region
    cropped_image = image[top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]]

    # Save the result images
    cv2.imwrite(output_path + "/matched_region.jpg", image)
    cv2.imwrite(output_path + "/cropped_template.jpg", cropped_image)

    print("Matching and cropping done. Results saved in", output_path)

# Paths
image_path = "traffic_image.jpg"  # Replace with the path to your input image
template_path = "template.jpg"   # Replace with the path to your template image
output_path = "."                # Replace with the desired output directory

# Call the function
template_matching_and_crop(image_path, template_path, output_path)
```
